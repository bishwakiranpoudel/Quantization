## 1. DOWNCASTING

```python
import torch

# random pytorch tensor: float32, size=1000
tensor_fp32 = torch.rand(1000, dtype = torch.float32)
print(tensor_fp32[:5])
```

```python
# downcast the tensor to bfloat16 using the "to" method
tensor_fp32_to_bf16 = tensor_fp32.to(dtype = torch.bfloat16)
print(tensor_fp32_to_bf16[:5])
```

```python
# tensor_fp32 x tensor_fp32
m_float32 = torch.dot(tensor_fp32, tensor_fp32)
print(m_float32)
```

```python
# tensor_fp32_to_bf16 x tensor_fp32_to_bf16
m_bfloat16 = torch.dot(tensor_fp32_to_bf16, tensor_fp32_to_bf16)
print(m_bfloat16)
```

## 2. LINEAR QUANTIZATION: MATRIX FORM

```python
import numpy as np
import matplotlib.pyplot as plt

# Original 3x3 matrix with high precision (floating-point numbers)
original_matrix = np.array([
    [0.123456789, 0.987654321, 0.456789123],
    [0.567891234, 0.234567891, 0.678912345],
    [0.345678912, 0.789123456, 0.912345678]
])

# Function to quantize the matrix to lower precision (e.g., INT8-like with rounding to 1 decimal place)
def quantize(matrix, decimals=1):
    return np.round(matrix, decimals=decimals)

# Quantized 3x3 matrix
quantized_matrix = quantize(original_matrix, decimals=1)

# Plotting the original and quantized matrices
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot original matrix
axes[0].imshow(original_matrix, cmap='coolwarm', interpolation='nearest')
axes[0].set_title('Original Matrix (High Precision)')
for i in range(original_matrix.shape[0]):
    for j in range(original_matrix.shape[1]):
        axes[0].text(j, i, f'{original_matrix[i, j]:.4f}', ha='center', va='center', color='white')

# Plot quantized matrix
axes[1].imshow(quantized_matrix, cmap='coolwarm', interpolation='nearest')
axes[1].set_title('Quantized Matrix (Lower Precision)')
for i in range(quantized_matrix.shape[0]):
    for j in range(quantized_matrix.shape[1]):
        axes[1].text(j, i, f'{quantized_matrix[i, j]:.1f}', ha='center', va='center', color='white')

plt.tight_layout()
axes[0].axis('off')
axes[1].axis('off')
plt.show()
```

```python
# Compute the quantization error matrix
quantization_error_matrix = original_matrix - quantized_matrix

# Plotting the quantization error matrix only
fig, ax = plt.subplots(figsize=(5, 5))

# Define the colormap
cmap = 'coolwarm'

# Plot quantization error matrix
cax = ax.imshow(quantization_error_matrix, cmap=cmap, interpolation='nearest')
ax.set_title('Quantization Error Matrix')
for i in range(quantization_error_matrix.shape[0]):
    for j in range(quantization_error_matrix.shape[1]):
        ax.text(j, i, f'{quantization_error_matrix[i, j]:.4f}', ha='center', va='center', color='white')

# Add color bar to show the scale
fig.colorbar(cax, ax=ax)

ax.axis('off')  # Corrected this line to use 'ax' instead of 'axes'

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Example weights
weights = np.array([[0.1, -0.3, 0.4],
                    [-0.2, 0.5, -0.1],
                    [0.2, -0.4, 0.3]])

# Quantization parameters
q_min, q_max = 0, 255

# Step 1: Calculate min and max of the weights
x_min = weights.min()
x_max = weights.max()

# Step 2: Calculate scale and zero-point
scale = (x_max - x_min) / (q_max - q_min)
zero_point = q_min - x_min / scale

# Step 3: Quantize the weights
weights_quantized = np.round(weights / scale + zero_point).astype(np.int32)

# Step 4: Dequantize to verify
weights_dequantized = scale * (weights_quantized - zero_point)

print("Original Weights:\n", weights)
print("Quantized Weights:\n", weights_quantized)
print("Dequantized Weights:\n", weights_dequantized)
```

#### Given:
- Floating-point range: $[x_{\text{min}}, x_{\text{max}}]$
- Quantized range: $[q_{\text{min}}, q_{\text{max}}]$

In this example, the quantization parameters $q_{\text{min}}$ and $q_{\text{max}}$ are set to 0 and 255, respectively, which correspond to quantization to an unsigned 8-bit integer (uint8) format. This meas that the range of possible quantized values is from 0 to 255, inclusive.

#### Formulas:
1. **Calculate Minimum and Maximum Values:**


   $[ x_{\text{min}} = \min(x), x_{\text{max}} = \max(x) ]$
   

   
   $[ q_{\text{min}} = 0, q_{\text{max}} = 255 ]$



2. **Compute Scale and Zero-Point:**


   $s = \frac{x_{\text{max}} - x_{\text{min}}}{q_{\text{max}} - q_{\text{min}}}$
   

   
   $z = q_{\text{min}} - \frac{x_{\text{min}}}{s}$


3. **Quantize the Values:**


   $q = \text{round}\left(\frac{x}{s} + z\right)$


4. **Dequantize to Verify:**


   $x_{\text{dequantized}} = s \cdot (q - z)$

```python
def named_module_tensors(module, recurse=False):
    for named_parameter in module.named_parameters(recurse=recurse):
      name, val = named_parameter
      flag = True
      if hasattr(val,"_data") or hasattr(val,"_scale"):
        if hasattr(val,"_data"):
          yield name + "._data", val._data
        if hasattr(val,"_scale"):
          yield name + "._scale", val._scale
      else:
        yield named_parameter

    for named_buffer in module.named_buffers(recurse=recurse):
      yield named_buffer

def dtype_byte_size(dtype):
    """
    Returns the size (in bytes) occupied by one parameter of type `dtype`.
    """
    import re
    if dtype == torch.bool:
        return 1 / 8
    bit_search = re.search(r"[^\d](\d+)$", str(dtype))
    if bit_search is None:
        raise ValueError(f"`dtype` is not a valid dtype: {dtype}.")
    bit_size = int(bit_search.groups()[0])
    return bit_size // 8
```

```python
def compute_module_sizes(model):
    """
    Compute the size of each submodule of a given model.
    """
    from collections import defaultdict
    module_sizes = defaultdict(int)
    for name, tensor in named_module_tensors(model, recurse=True):
      size = tensor.numel() * dtype_byte_size(tensor.dtype)
      name_parts = name.split(".")
      for idx in range(len(name_parts) + 1):
        module_sizes[".".join(name_parts[:idx])] += size

    return module_sizes
```

```python
import torch

module_sizes = compute_module_sizes(model)
print(f"The model size is {module_sizes[''] * 1e-9} GB")
```
